```python
import pickle
```

```python
pickled_data = pickle.load(open('UMAP_benchmark_results_feb/benchmark_umap_embeddings_big.pkl', 'rb'))
```

```python
embedding = pickled_data.loc[1]['embedding']
embedding.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import umap
%matplotlib inline
sns.set(style='white', rc={'figure.figsize':(12,8)})

import datashader as ds
import datashader.transfer_functions as tf
import colorcet
import pandas as pd
```

```python
cvs = ds.Canvas(plot_width=800, plot_height=600)
agg = cvs.points(pd.DataFrame(embedding, columns=['x_col', 'y_col']), 'x_col', 'y_col')
img = tf.set_background(tf.interpolate(agg, cmap=colorcet.fire, how='log'),"black")
img
```
